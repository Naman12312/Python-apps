```python
from matplotlib import style
from matplotlib import pyplot as plt
import numpy as np
style.use("dark_background")
```

```python
x = np.array([23,44,61])
y = np.array([24, 43, 87])
x2 = np.array([33,56,21])
y2 = np.array([21, 44, 67])
```

```python
print(x)
plt.title("Matplotlib Tutorial")
plt.xlabel("X")
plt.ylabel("Y")
plt.plot(x, y, label = "X and Y", linewidth=4)
plt.plot(x2, y2, label = "X2 and Y2")
plt.legend()

```

```python
plt.show()
```
