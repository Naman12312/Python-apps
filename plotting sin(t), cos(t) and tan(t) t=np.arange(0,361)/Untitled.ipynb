```python
import pandas as pd
import numpy as np

from decimal import Decimal as d
from matplotlib import pyplot as plt
my_arr= np.radians(np.arange(0,361))

```

```python
df = pd.DataFrame({
    "sin":np.array(np.sin(np.deg2rad(np.arange(0,361)))),
    "cos":np.array(np.cos(np.deg2rad(np.arange(0,361)))),
    "tan":np.array(np.tan(np.deg2rad(np.arange(0,361)))),
}, index=np.arange(0, 361))
```
