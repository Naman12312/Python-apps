```python
from matplotlib import pyplot as plt
from matplotlib import style
import numpy as np
```

```python
x = [11,51,41]
y = [23,45,89]
x2 = np.array([15,7,3])
y2 = [5,8,9]
# exp = np.zeros(6) 
# beats = np.array([180, 33, 46, 76, 88, 222])
exp = np.zeros(len(x2))
exp[x2.argmax()] = 0.1
a = ['Naman','Mayank','Punya']
style.use("ggplot")
plt.plot(x,y,x2,y2)
plt.xlabel = "X"
plt.ylabel = "Y"
plt.legend(["XY", "X2Y2"])
plt.show()
```

```python
x = [11,51,41]
y = [23,45,89]
x2 = np.array([15,7,3])
y2 = [5,8,9]
# exp = np.zeros(6) 
# beats = np.array([180, 33, 46, 76, 88, 222])
exp = np.zeros(len(x2))
exp[x2.argmax()] = 0.1
a = ['Naman','Mayank','Punya']
style.use("ggplot")
plt.pie(x2 ,labels = a, explode = exp, colors = ['red', 'green', 'blue'])
plt.legend()
plt.show()
```

# Setting range(max_limit) for x and y axis

```python
from matplotlib import pyplot as plt

fig, ax = plt.subplots()
ax.set_ylim([0,500]) # setting limit for y axis
ax.set_xlim([0,100]) # setting limit for x axis
plt.bar(height=[100,50,200], x=[1,2,4])
plt.show()

```
