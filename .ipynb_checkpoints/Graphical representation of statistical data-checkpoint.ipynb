```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Define your data
categories = ['Category 1', 'Category 2', 'Category 3']
values = [10, 15, 7]

# Create a new figure
fig, ax = plt.subplots()

# For each category and its corresponding value
for i, (category, value) in enumerate(zip(categories, values)):
    # Plot a number of icons equal to the value
    for x in range(value):
        # We'll use circles for the icons, but you can use other shapes or even custom icons
        circle = mpatches.Circle((x + 1, i + 1), 0.4, facecolor='blue')
        ax.add_patch(circle)

# Set the limits and labels of the plot
ax.set_xlim([0, max(values) + 2])
ax.set_ylim([0, len(categories) + 2])
ax.set_yticks(range(1, len(categories) + 1))
ax.set_yticklabels(categories)

# Show the plot
plt.show()
```
