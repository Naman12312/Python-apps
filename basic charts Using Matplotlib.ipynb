# Setting range(max_limit) for x and y axis

**This isn't required most of the times, but just in case, if you need it, here it is:**

```python
from matplotlib import pyplot as plt

fig, ax = plt.subplots()
ax.set_ylim([0,300]) # setting limit for y axis
ax.set_xlim([0,50]) # setting limit for x axis
plt.bar([1,2,4],[100,50,200], color=("red", "green", "blue"))
plt.show()

```
